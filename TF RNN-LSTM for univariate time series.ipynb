```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

tf.random.set_random_seed(1001)
```

## Dataset Loading

### USD_INR Exchange data

```python
d = pd.read_csv("../Python/Datasets/USD_INR.csv")['Price']
d = pd.concat([d, d.shift(-1), d.shift(-2), d.shift(-3), d.shift(-4), d.shift(-5)], axis=1).iloc[:-5]
d.columns = ['t-5', 't-4', 't-3', 't-2', 't-1', 'out']

split = int(0.7*d.shape[0])
```

### Shampoo sales data

### Airpassengers data

```python
input_train, output_train = d.values[:split,:-1].astype(np.float64), d.values[:split, -1:].astype(np.float64)
input_test, output_test = d.values[split:,:-1].astype(np.float64), d.values[split:, -1:].astype(np.float64)
```

## Model Creation

```python
class LSTM_Model:
    
    def __init__(self, out_dim, fc_dim, rnn_dim, seq_length):
        
        self.out_dim = out_dim
        self.fc_dim = fc_dim
        self.n_fc = len(fc_dim)
        self.rnn_dim = rnn_dim
        self.seq_length = seq_length
        
    def initialize_weights(self):
        
        W,B = [],[]
        with tf.variable_scope("weights"):
            if(len(self.rnn_dim)==1):
                rnn_cell = tf.contrib.rnn.LSTMCell(self.rnn_dim[0])
            else:
                rnn_cell = tf.contrib.rnn.MultiRNNCell([tf.contrib.rnn.LSTMCell(i) for i in self.rnn_dim])
            if(len(self.fc_dim) == 0):
                W.append(tf.get_variable("w0", 
                                         shape=(self.rnn_dim[-1],self.out_dim),
                                         initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                         dtype=tf.float64))
                B.append(tf.get_variable("b0", 
                                         shape=(1,self.out_dim),
                                         initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                         dtype=tf.float64))
            else:
                for i in range(len(self.fc_dim)+1):
                    if(i==0):
                        W.append(tf.get_variable("w"+str(i), 
                                                 shape=(self.rnn_dim[-1],self.fc_dim[i]),
                                                 initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                                 dtype=tf.float64))
                        B.append(tf.get_variable("b"+str(i), 
                                                 shape=(1,self.fc_dim[i]),
                                                 initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                                 dtype=tf.float64))

                    elif(i==self.n_fc):
                        W.append(tf.get_variable("w"+str(i), 
                                                 shape=(self.fc_dim[i-1],self.out_dim), 
                                                 initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                                 dtype=tf.float64))
                        B.append(tf.get_variable("b"+str(i), 
                                                 shape=(1,self.out_dim), 
                                                 initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                                 dtype=tf.float64))

                    else:
                        W.append(tf.get_variable("w"+str(i), 
                                                 shape=(self.fc_dim[i-1],self.fc_dim[i]), 
                                                 initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                                 dtype=tf.float64))
                        B.append(tf.get_variable("b"+str(i), 
                                                 shape=(1,self.fc_dim[i]), 
                                                 initializer=tf.glorot_normal_initializer(dtype=tf.float64), 
                                                 dtype=tf.float64))
        return rnn_cell, W, B
    
    def make_flow_graph(self, x, y):
        
        with tf.variable_scope("weights", reuse=tf.AUTO_REUSE):
            
            rnn_cell, W, B = self.initialize_weights()
            
            x = tf.reshape(x, [-1,self.seq_length])
            x = tf.split(x, self.seq_length, 1)
            
            outputs, states = tf.contrib.rnn.static_rnn(cell=rnn_cell, 
                                                        inputs=x, 
                                                        dtype=tf.float64)

            tf.summary.histogram("LSTM_OUT",outputs[-1])
            
            l = outputs[-1]

            for i in range(len(W)-1):
                l = tf.nn.relu(tf.add(tf.matmul(l, W[i]), B[i]))
            
            out = tf.nn.relu(tf.add(tf.matmul(l,W[-1]), B[-1]))
            
            tf.summary.histogram("Predictions", out)

            for i in range(len(W)):
                tf.summary.histogram("w"+str(i), W[i])
            
            loss = tf.reduce_mean(tf.squared_difference(out,y))
            optim = tf.train.AdamOptimizer(learning_rate=0.0001).minimize(loss)

            with tf.variable_scope("metrics", reuse=tf.AUTO_REUSE):
                total_error = tf.reduce_sum(tf.square(tf.subtract(y, tf.reduce_mean(y))))
                unexplained_error = tf.reduce_sum(tf.square(tf.subtract(y, out)))
                R_squared = tf.subtract(tf.cast(1.0, dtype=tf.float64), tf.divide(unexplained_error, total_error))

                mae = tf.reduce_mean(tf.abs(tf.subtract(out,y)))

            tf.summary.scalar("Loss",loss)
            tf.summary.scalar("R2",R_squared)
            tf.summary.scalar("MAE",mae)

            return loss, optim, R_squared, mae, out
    
    def model_run(self, input_train, output_train, input_test, output_test, epochs, batch_size):
        
        with tf.Session() as sess:
            
            x = tf.placeholder(dtype=tf.float64, shape=(None, self.seq_length))
            y = tf.placeholder(dtype=tf.float64, shape=(None, 1))
            
            loss, optim, r2, mae, out = self.make_flow_graph(x,y)
            
            merged = tf.summary.merge_all()
            writer = tf.summary.FileWriter("./logs/logs3/run10/",sess.graph)
            
            sess.run(tf.global_variables_initializer())
            cnt = -1
            
            for ep in range(epochs):
                print("Epoch : ", ep)
                for mini in range(0,input_train.shape[0],batch_size):
                    r, m, s, l, _, o = sess.run([r2, mae, merged, loss, optim, out], feed_dict={x:input_train[mini:mini+batch_size], y:output_train[mini:mini+batch_size]})
                    print("Batch no : %d \tLoss : %0.5f \tR2 : %0.3f \tMAE : %0.3f" %(int(mini/batch_size), np.mean(l), r, m))
                    cnt+=1
                    writer.add_summary(s, cnt)
                print("------------------------------------------------------------------------------------")
            
            r, m, l, o = sess.run([r2, mae, loss, out], feed_dict={x:input_test, y:output_test})
            print("Loss : %0.3f \tR2 : %0.3f \tMAE : %0.3f" %(np.mean(l), np.mean(r), np.mean(m)))
            
            return o
            
```

```python
model = LSTM_Model(out_dim=1, fc_dim=[], rnn_dim=[64,64,32], seq_length=input_train.shape[1])
```

```python
preds = model.model_run(input_train=input_train, 
                output_train=output_train, 
                input_test=input_test, 
                output_test=output_test, 
                epochs=100, 
                batch_size=16)
```

```python
p = preds
gt = output_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(mean_squared_error(p,gt), mean_absolute_error(p,gt), r2_score(p,gt))
```
