```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Building a simple ANN with 2 hidden layers

```python
inp = tf.placeholder(tf.float32, shape=(None,1))
out = tf.placeholder(tf.float32, shape=(None,))
```

```python
with tf.variable_scope("params", reuse=tf.AUTO_REUSE):
    w1 = tf.get_variable("w1", shape=(1,4), initializer=tf.glorot_normal_initializer(seed=92, dtype=tf.float32))
    b1 = tf.get_variable("b1", shape=(4), initializer=tf.glorot_normal_initializer(seed=92, dtype=tf.float32))
    
    w2 = tf.get_variable("w2", shape=(4,8), initializer=tf.glorot_normal_initializer(seed=92, dtype=tf.float32))
    b2 = tf.get_variable("b2", shape=(8), initializer=tf.glorot_normal_initializer(seed=92, dtype=tf.float32))
    
    w3 = tf.get_variable("w3", shape=(8,1), initializer=tf.glorot_normal_initializer(seed=92, dtype=tf.float32))
    b3 = tf.get_variable("b3", shape=(1), initializer=tf.glorot_normal_initializer(seed=92, dtype=tf.float32))
    
with tf.variable_scope("calc", reuse=tf.AUTO_REUSE):
    l1 = tf.nn.relu6(tf.add(tf.matmul(inp,w1),b1))
    l2 = tf.nn.relu6(tf.add(tf.matmul(l1,w2),b2))
    lo = tf.add(tf.matmul(l2,w3),b3)
    
loss = tf.reduce_mean(tf.square(out-lo))
step = tf.train.AdamOptimizer().minimize(loss)
```

```python
x = np.linspace(0,1,5000).reshape((-1,1))
y = (5*x + 9).reshape((-1,))
```

```python
x_train, x_test = x[:4000], x[4000:]
y_train, y_test = y[:4000], y[4000:]
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
epochs=10
batch_size=32
```

```python
plt.plot(x_train,y_train)
plt.plot(x_test,y_test)
plt.show()
```

```python
for e in range(epochs):
    for b in range(0,x_train.shape[0],batch_size):
        l,_=sess.run([loss,step], {inp: x_train[b:b+batch_size], out: y_train[b:b+batch_size]})
        print("Training Epoch {0:2d} Batch {1:5d} Loss {2:5.3f}".format(e+1,(b//batch_size)+1,l))
        
    outs = []
    for b in range(0,x_test.shape[0],batch_size):
        l,ot=sess.run([loss,lo], {inp: x_test[b:b+batch_size], out: y_test[b:b+batch_size]})
        print("Validation Batch {0:2d} Loss {1:5.3f}".format((b//batch_size)+1,l))
        outs.extend(ot.tolist())
    plt.figure()
    plt.plot(x_test,y_test)
    plt.plot(x_test,outs)
    plt.show()
```
